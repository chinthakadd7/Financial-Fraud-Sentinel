### Train H2O AutoML

```python
import h2o
h2o.init(max_mem_size="8G", nthreads=4)
```

```python
DATA_PATH = "/content/drive/MyDrive/financial-fraud-sentinel/data/processed/fraud_processed.csv"

hf = h2o.import_file(DATA_PATH)
```

```python
hf["isFraud"] = hf["isFraud"].asfactor()
```

```python
from h2o.automl import H2OAutoML

x=hf.columns
y="isFraud"
x.remove(y)

aml = H2OAutoML(
    max_runtime_secs =900,
    max_models =15,
    balance_classes=True
)

aml.train(x=x, y=y,training_frame=hf)
```

```python
leader = aml.leader
```

```python
leader.download_mojo(
    path="/content/drive/MyDrive/financial-fraud-sentinel/models",
    get_genmodel_jar = True
)
```
