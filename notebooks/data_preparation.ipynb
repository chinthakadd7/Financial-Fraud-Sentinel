### Start Working on Data

```python
import pandas as pd

DATA_PATH = "/content/drive/MyDrive/financial-fraud-sentinel/data/raw"

transaction = pd.read_csv(f"{DATA_PATH}/train_transaction.csv")
identity = pd.read_csv(f"{DATA_PATH}/train_identity.csv")
```

```python
df = transaction.merge(identity, on="TransactionID", how ="left")
```

clear missing values

```python
for col in df.columns:
  if df[col].dtype == "object":
    df[col] = df[col].fillna("Unknown")
  else:
    df[col] = df[col].fillna(df[col].median())
```

Feature Engineering

```python
df["is_large_amount"] = (df["TransactionAmt"]> 1000).astype(int)
df["transaction_count"] = df.groupby("card1")["TransactionID"].transform("count")
```

```python
PROCESSED_PATH = "/content/drive/MyDrive/financial-fraud-sentinel/data/processed"
df.to_csv(f"{PROCESSED_PATH}/fraud_processed.csv", index=False)
```
